# Data Visualization Using Seaborn Library
# Plotting with categorical data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

## Import different dataset

```python
titanic_dataset = sns.load_dataset("titanic")
tips_dataset = sns.load_dataset("tips")
iris_dataset = sns.load_dataset("iris")
```

## Categorical scatterplots

```python
sns.catplot(x = "day", y = "total_bill", data = tips_dataset)
```

```python
sns.catplot(x = "day", y = "total_bill", jitter = False, data = tips_dataset)
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "swarm", data = tips_dataset)
```

```python
sns.catplot(x = "day", y = "total_bill", hue = "sex", kind = "swarm", data = tips_dataset)
```

## Distributions of observations within categories
### Boxplots

```python
sns.catplot(x = "day", y = "total_bill", kind = "box", data = tips_dataset)
```

## Violinplots

```python
sns.catplot(x = "total_bill", y = "day", hue = "sex",
            kind = "violin", data = tips_dataset)
```

```python
sns.catplot(x = "day", y = "total_bill", hue = "sex",
            kind = "violin", split = True, data = tips_dataset)
```

## Plotting wide-form data

```python
sns.catplot(data = iris_dataset, orient = "h", kind = "box")
```

```python
sns.violinplot(x = iris_dataset.species, y = iris_dataset.sepal_length)
```
