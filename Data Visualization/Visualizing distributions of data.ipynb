# Data Visualization Using Seaborn Library
# Visualizing distributions of data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

## Import different datasets

```python
penguins_dataset = sns.load_dataset("penguins")
tips_dataset = sns.load_dataset("tips")
```

```python
penguins_dataset.head()
```

```python
tips_dataset.head()
```

## Plotting univariate histograms

```python
sns.displot(x = "flipper_length_mm", data = penguins_dataset)
```

## Choosing the bin size

```python
sns.displot(x = "flipper_length_mm", binwidth = 3,data = penguins_dataset)
```

#### _In other circumstances, it may make more sense to specify the number of bins, rather than their size:_

```python
sns.displot(x = "flipper_length_mm", bins = 20, data = penguins_dataset)
```

## Conditioning on other variables

```python
sns.displot(x = "flipper_length_mm", hue = "species", data = penguins_dataset)
```

```python
sns.displot(x = "flipper_length_mm", hue = "species", element = "step", data = penguins_dataset)
```

#### _Alternatively, instead of layering each bar, they can be stacked, or moved vertically. In this plot, the outline of the full histogram will match the plot with only a single variable:_

```python
sns.displot(x = "flipper_length_mm", hue = "species", multiple = "stack", data = penguins_dataset)
```

```python
sns.displot(penguins_dataset, x="flipper_length_mm", hue="sex", multiple="dodge")
```

```python
sns.displot(penguins_dataset, x="flipper_length_mm", col="sex")
```

## Kernel density estimation

```python
sns.displot(penguins_dataset, x = "flipper_length_mm", kind = "kde")
```

## Choosing the smoothing bandwidth

```python
sns.displot(penguins_dataset, x="flipper_length_mm", kind="kde", bw_adjust=.25)
```

```python
sns.displot(penguins_dataset, x="flipper_length_mm", kind="kde", bw_adjust=2)
```

## Conditioning on other variables

```python
sns.displot(penguins_dataset, x="flipper_length_mm", hue="species", kind="kde")
```

```python
sns.displot(penguins_dataset, x="flipper_length_mm", hue="species", kind="kde", multiple="stack")
```

```python
sns.displot(penguins_dataset, x="flipper_length_mm", hue="species", kind="kde", fill=True)
```

## Kernel density estimation pitfalls

```python
sns.displot(tips_dataset, x = "total_bill", kind = "kde")
```

## Empirical cumulative distributions

```python
sns.displot(penguins_dataset, x = "flipper_length_mm", kind = "ecdf")
```

```python
sns.displot(penguins_dataset, x = "flipper_length_mm", hue = "species", kind = "ecdf")
```

## Visualizing bivariate distributions

```python
sns.displot(penguins_dataset, x = "bill_length_mm", y = "bill_depth_mm")
```

```python
sns.displot(penguins_dataset, x = "bill_length_mm", y = "bill_depth_mm", kind = "kde")
```

```python
sns.displot(penguins_dataset, x = "bill_length_mm", y = "bill_depth_mm", hue = "species")
```

```python
sns.displot(penguins_dataset, x = "bill_length_mm", y = "bill_depth_mm", hue = "species", kind = "kde")
```

```python
sns.displot(penguins_dataset, x = "bill_length_mm", y = "bill_depth_mm", cbar=True)
```

## Plotting joint and marginal distributions

```python
sns.jointplot(data = penguins_dataset, x = "bill_length_mm", y = "bill_depth_mm")
```

```python
sns.jointplot(
    data=penguins_dataset,
    x = "bill_length_mm", y = "bill_depth_mm", hue = "species",
    kind = "kde"
)
```

```python
sns.displot(
    penguins_dataset, x = "bill_length_mm", y = "bill_depth_mm",
    kind = "kde", rug = True
)
```

```python
sns.relplot(data = penguins_dataset, x = "bill_length_mm", y = "bill_depth_mm")
sns.rugplot(data = penguins_dataset, x = "bill_length_mm", y = "bill_depth_mm")
```

```python
sns.pairplot(penguins_dataset)
```

```python
g = sns.PairGrid(penguins_dataset)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill = True)
g.map_diag(sns.histplot, kde = True)
```
