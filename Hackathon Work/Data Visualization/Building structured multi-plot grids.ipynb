# Data Visualization Using Seaborn Library
## Building structured multi-plot grids
## Plotting pairwise data relationships

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

## Import different dataset

```python
titanic_dataset = sns.load_dataset("titanic")
tips_dataset = sns.load_dataset("tips")
iris_dataset = sns.load_dataset("iris")
```

```python
g = sns.PairGrid(iris_dataset)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(iris_dataset)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(iris_dataset, hue = "species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
g = sns.PairGrid(iris_dataset)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw = 3, legend = False)
```

```python
sns.pairplot(iris_dataset, hue = "species", height = 2.5)
```

```python
g = sns.pairplot(iris_dataset, hue = "species", palette = "Set2", diag_kind = "kde", height = 2.5)
```
